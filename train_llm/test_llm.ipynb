```python
from transformers import AutoTokenizer, AutoModelForCausalLM, AutoConfig
import torch
from train import LLM, Config
```

```python
t = AutoTokenizer.from_pretrained('/root/autodl-tmp/train_llm_from_scratch/saves/model')
AutoConfig.register("small_model", Config)
AutoModelForCausalLM.register(Config, LLM)
```

```python
model = AutoModelForCausalLM.from_pretrained('/root/autodl-tmp/train_llm_from_scratch/saves/model')
```

```python
input_data = [t.bos_token_id] + t.encode('13+28等于')
print(input_data)
```

```python
for token in model.generate({"input_ids":torch.tensor(input_data).unsqueeze(0), "labels":None}, t.eos_token_id, 20, stream=False,temperature=0.0, top_k=8):
    print(t.decode(token[0]))
```
